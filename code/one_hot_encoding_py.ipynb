```python
# creating one hot encoding
import numpy as np
import pandas as pd
import re
```

```python
# pre processing
def quick_preprocess(text):
    text = 'NULL' if text is None else str(text)
    x = text.lower().replace("."," ").replace("."," ").replace("\\n", " ")
    return_text = re.sub(' +', ' ', str(x))
    return return_text
```

```python
input_file = "/home/sandipan/Insync/ghoshm21@gmail.com/Google Drive/personal_project/Learning_NLP/data/input/sample_text.txt"
data = pd.read_csv(input_file,sep="\n",header=None)
data.columns = ['text_data']
print(data)
# apply the pre-process
data["text_data"] = data["text_data"].apply(quick_preprocess)
print(data)
# convert the docuemnts to list of unique words
processed_docs =  list(data['text_data'].str.split(' ', expand=True).stack().unique())
print(processed_docs)
# # read the file
# f = open(input_file, "r")
# documents = f.read()
# # # print(documents)
# # # convert the docuemnts to list of unique words
# # processed_docs = np.unique(np.array(documents.lower().replace("."," ").replace("\\n", " ").split()))

# # print(processed_docs)
```

```python
#Build the vocabulary
vocab = {}
for x in range (len(processed_docs)):
    # print(x+1)
    vocab[processed_docs[x]] = x+1

print(vocab)
```

```python
# function for one hot encoding
def get_onehot_encoding(input):
    onehot_encode = []
    for word in input.split():
        temp_encode = [0]*len(vocab)
        if word in vocab:
            temp_encode[vocab[word]-1] = 1 # -1 is to take care of the fact indexing in array starts from 0 and not 1
        onehot_encode.append(temp_encode)
    return onehot_encode
```
